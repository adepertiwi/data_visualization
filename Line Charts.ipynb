LINE CHARTS

Apa itu Line Chart?

Line chart adalah jenis grafik yang digunakan untuk menampilkan data dalam bentuk garis yang menghubungkan titik data. Grafik ini sangat berguna untuk menunjukkan tren atau pola dalam data seiring waktu atau urutan tertentu. Biasanya, sumbu horizontal (X) mewakili variabel independen, seperti waktu atau kategori, sedangkan sumbu vertikal (Y) mewakili variabel dependen, seperti jumlah, persentase, atau nilai lainnya.

Kapan Line Chart Digunakan?

Line chart digunakan ketika ingin menampilkan perubahan data sepanjang waktu atau urutan tertentu. Ini cocok untuk menunjukkan tren jangka panjang, pola musiman, atau fluktuasi data seiring waktu. Line chart juga efektif untuk membandingkan beberapa set data dan menyoroti perbedaan atau kesamaan pola di antara mereka.

Cara Membaca Line Chart:

Tren: Line chart membantu mengidentifikasi tren data seiring waktu. Jika garis naik, itu menunjukkan peningkatan nilai variabel dependen. Sebaliknya, jika garis turun, itu menunjukkan penurunan nilai. Garis horizontal menunjukkan stabilitas atau sedikit perubahan.

Polapola: Pola atau siklus yang berulang dapat ditemukan dalam line chart, yang membantu memahami fluktuasi data. Misalnya, dalam data penjualan, mungkin akan ada pola musiman di mana penjualan meningkat selama musim liburan.

Perbandingan: Line chart memungkinkan perbandingan antara beberapa set data. Ini berguna untuk melihat bagaimana pola atau tren berbeda antara beberapa variabel atau kelompok.

Outliers: Line chart dapat membantu mengidentifikasi titik-titik yang jauh dari pola umum, yang dapat menunjukkan kejadian luar biasa atau kesalahan dalam data.

Code menggunakan library Matplotlib untuk membuat line chart yang menyoroti tren data

```python
import matplotlib.pyplot as plt

# Data untuk line chart
tahun = [2015, 2016, 2017, 2018, 2019, 2020]
pendapatan_perusahaan_A = [50000, 55000, 60000, 62000, 65000, 70000]

# Membuat line chart
plt.figure(figsize=(10, 6))

# Plot tren pendapatan perusahaan A
plt.plot(tahun, pendapatan_perusahaan_A, marker='o', linestyle='-')

# Menambahkan judul dan label sumbu
plt.title('Tren Pendapatan Perusahaan A Tahun 2015-2020')
plt.xlabel('Tahun')
plt.ylabel('Pendapatan (dalam USD)')

# Menampilkan grid
plt.grid(True)

# Menampilkan grafik
plt.show()
```

Penjelasan Kodingan:

1. "import matplotlib.pyplot as plt" merupakan code yang digunakan untuk mengimpor library "matplotlib.pyplot" dan menggantikan namanya dengan "plt", sehingga kita dapat menggunakan fungsi-fungsi dari library ini dengan menggunakan "plt".

2. "tahun" dan "pendapatan_perusahaan_A" merupakan data yang akan digunakan untuk membuat line chart. "tahun" adalah daftar tahun dari 2015 hingga 2020, sedangkan "pendapatan_perusahaan_A" adalah daftar pendapatan perusahaan A dalam rentang waktu yang sama.

3. "plt.figure(figsize=(10, 6))" merupakan fungsi yang digunakan untuk membuat gambar baru dengan ukuran 10x6 inci.

4. "plt.plot(tahun, pendapatan_perusahaan_A, marker='o', linestyle='-')" merupakan perintah untuk membuat line chart dengan tahun di sumbu "x" dan pendapatan perusahaan A di sumbu "y". Argumen "marker='o'" berfungsi untuk membuat titik data agar ditandai dengan simbol lingkaran, sedangkan "linestyle='-'" untuk garis agar digambar dengan garis lurus.

5. "plt.title('Tren Pendapatan Perusahaan A Tahun 2015-2020')" untuk menambahkan judul dari line chart.

6. "plt.xlabel('Tahun')" dan "plt.ylabel('Pendapatan (dalam USD)')" berfungsi menambahkan label untuk sumbu x (tahun) dan sumbu y (pendapatan dalam USD).

7. "plt.grid(True)" untuk menampilkan grid di dalam line chart.

8. plt.show()" untuk menampilkan grafik line chart yang telah dibuat.

Dengan menjalankan kodingan ini, Anda akan mendapatkan line chart yang menunjukkan tren pendapatan perusahaan A dari tahun 2015 hingga 2020.

Code menggunakan library Matplotlib untuk membuat line chart yang menunjukkan pola pola tertentu

```python
# Data untuk line chart
bulan = ['Jan', 'Feb', 'Mar', 'Apr', 'Mei', 'Jun', 'Jul', 'Ags', 'Sep', 'Okt', 'Nov', 'Des']
penjualan = [100, 110, 120, 140, 130, 160, 170, 180, 190, 200, 220, 230]

# Membuat line chart
plt.figure(figsize=(10, 6))

# Plot data penjualan
plt.plot(bulan, penjualan, marker='o', linestyle='-')

# Menambahkan judul dan label sumbu
plt.title('Pola Penjualan Bulanan Tahun Ini')
plt.xlabel('Bulan')
plt.ylabel('Penjualan (dalam ribu unit)')

# Menampilkan grid
plt.grid(True)

# Menampilkan grafik
plt.show()
```

Penjelasan Kodingan:

1. "bulan" dan "penjualan" adalah data yang akan digunakan untuk membuat line chart. "bulan" adalah daftar bulan dari Januari hingga Desember, sedangkan "penjualan" adalah daftar jumlah penjualan dalam ribu unit untuk setiap bulan tersebut.

2. "plt.figure(figsize=(10, 6))" merupakan fungsi yang digunakan untuk membuat gambar baru dengan ukuran 10x6 inci.

3. "plt.plot(bulan, penjualan, marker='o', linestyle='-')" adalah perintah untuk membuat line chart dengan bulan di sumbu x dan jumlah penjualan di sumbu y. Argumen "marker='o'" menandakan bahwa titik data akan ditandai dengan simbol lingkaran, sedangkan "linestyle='-'" menunjukkan bahwa garis akan digambar dengan garis lurus.

4. "plt.title('Pola Penjualan Bulanan Tahun Ini')" berfungsi untuk menambahkan judul untuk line chart.

5. "plt.xlabel('Bulan')" dan "plt.ylabel('Penjualan (dalam ribu unit)')" untuk menambahkan label untuk sumbu x (bulan) dan sumbu y (penjualan dalam ribu unit).

6. "plt.grid(True)" untuk menampilkan grid di dalam line chart.

7. "plt.show()" untuk menampilkan grafik line chart yang telah dibuat.

Dengan menjalankan kodingan ini, akan didapatkan line chart yang menunjukkan pola penjualan bulanan untuk tahun ini, mulai dari Januari hingga Desember.

Code menggunakan library Matplotlib untuk membuat line chart yang membandingkan dua set data



```python
# Data untuk line chart
tahun = [2015, 2016, 2017, 2018, 2019, 2020]
pendapatan_perusahaan_A = [50000, 55000, 60000, 62000, 65000, 70000]
pendapatan_perusahaan_B = [45000, 52000, 58000, 61000, 63000, 69000]

# Membuat line chart
plt.figure(figsize=(10, 6))

# Plot tren pendapatan perusahaan A
plt.plot(tahun, pendapatan_perusahaan_A, marker='o', linestyle='-', label='Perusahaan A')

# Plot tren pendapatan perusahaan B
plt.plot(tahun, pendapatan_perusahaan_B, marker='s', linestyle='--', label='Perusahaan B')

# Menambahkan judul dan label sumbu
plt.title('Perbandingan Pendapatan Perusahaan A dan B Tahun 2015-2020')
plt.xlabel('Tahun')
plt.ylabel('Pendapatan (dalam USD)')

# Menampilkan legend
plt.legend()

# Menampilkan grid
plt.grid(True)

# Menampilkan grafik
plt.show()
```

Penjelasan Kodingan:

1. "tahun", "pendapatan_perusahaan_A", dan "pendapatan_perusahaan_B" merupakan data yang akan digunakan untuk membuat line chart. "tahun" adalah daftar tahun dari 2015 hingga 2020, "pendapatan_perusahaan_A" adalah daftar pendapatan perusahaan A dalam rentang waktu yang sama, dan "pendapatan_perusahaan_B" adalah daftar pendapatan perusahaan B dalam rentang waktu yang sama.

2. "plt.figure(figsize=(10, 6))" merupakan fungsi ini digunakan untuk membuat gambar baru dengan ukuran 10x6 inci.

3. "plt.plot(tahun, pendapatan_perusahaan_A, marker='o', linestyle='-', label='Perusahaan A')" adalah perintah untuk membuat line chart dengan tahun di sumbu x dan pendapatan perusahaan A di sumbu y. Argumen "marker='o'" menandakan bahwa titik data akan ditandai dengan simbol lingkaran, sedangkan "linestyle='-'" menunjukkan bahwa garis akan digambar dengan garis lurus. Argumen "label='Perusahaan A'" menambahkan label untuk line chart tersebut.

4. "plt.plot(tahun, pendapatan_perusahaan_B, marker='s', linestyle='--', label='Perusahaan B')" adalah perintah untuk membuat line chart dengan tahun di sumbu x dan pendapatan perusahaan B di sumbu y. Argumen "marker='s'" menandakan bahwa titik data akan ditandai dengan simbol persegi, sedangkan "linestyle='--'" menunjukkan bahwa garis akan digambar dengan garis putus-putus. Argumen "label='Perusahaan B'" menambahkan label untuk line chart tersebut.

5. "plt.title('Perbandingan Pendapatan Perusahaan A dan B Tahun 2015-2020')" berfungsi untuk menambahkan judul untuk line chart.

6. "plt.xlabel('Tahun')" dan "plt.ylabel('Pendapatan (dalam USD)')" berfungsi untuk menambahkan label untuk sumbu x (tahun) dan sumbu y (pendapatan dalam USD).

7. "plt.legend()" berfungsi untuk menampilkan legenda yang menjelaskan garis-garis dalam grafik, yaitu perbandingan antara pendapatan perusahaan A dan B.

8. "plt.grid(True)" berfungsi untuk menampilkan grid di dalam line chart.

9. "plt.show()" berfungsi untuk menampilkan grafik line chart yang telah dibuat.

Code Python menggunakan library Matplotlib untuk membuat line chart yang menyoroti outliers



```python
# Data untuk line chart
tahun = [2015, 2016, 2017, 2018, 2019, 2020]
pendapatan = [50000, 55000, 60000, 62000, 65000, 70000]
outliers = [51000, 70000, 40000, 75000, 73000, 68000]

# Membuat line chart
plt.figure(figsize=(10, 6))

# Plot tren pendapatan
plt.plot(tahun, pendapatan, marker='o', linestyle='-', label='Pendapatan')

# Plot outliers
plt.scatter(tahun, outliers, color='red', label='Outliers')

# Menambahkan judul dan label sumbu
plt.title('Pendapatan Tahunan dengan Outliers')
plt.xlabel('Tahun')
plt.ylabel('Pendapatan (dalam USD)')

# Menampilkan legend
plt.legend()

# Menampilkan grid
plt.grid(True)

# Menampilkan grafik
plt.show()
```

Penjelasan Kodingan:

1. "tahun", "pendapatan", dan "outliers" adalah data yang akan digunakan untuk membuat line chart. "tahun" adalah daftar tahun dari 2015 hingga 2020, "pendapatan" adalah daftar pendapatan tahunan dalam rentang waktu yang sama, dan "outliers" adalah daftar nilai-nilai yang dianggap sebagai outliers atau nilai-nilai yang jauh dari nilai-nilai lainnya.

2. "plt.figure(figsize=(10, 6))" merupakan fungsi ini digunakan untuk membuat gambar baru dengan ukuran 10x6 inci.

3. "plt.plot(tahun, pendapatan, marker='o', linestyle='-', label='Pendapatan')" merupakan perintah untuk membuat line chart dengan tahun di sumbu x dan pendapatan di sumbu y. Argumen "marker='o'" menandakan bahwa titik data akan ditandai dengan simbol lingkaran, sedangkan "linestyle='-'" menunjukkan bahwa garis akan digambar dengan garis lurus. Argumen "label='Pendapatan'" menambahkan label untuk line chart tersebut.

4. "plt.scatter(tahun, outliers, color='red', label='Outliers')" adalah perintah untuk menampilkan outliers sebagai titik-titik berwarna merah di atas line chart. Fungsi "scatter" digunakan untuk menampilkan titik-titik individual. Argumen "color='red'" menandakan bahwa titik-titik outliers akan ditampilkan dengan warna merah. Argumen "label='Outliers'" menambahkan label untuk outliers.

5. "plt.title('Pendapatan Tahunan dengan Outliers')" berfungsi untuk menambahkan judul untuk line chart.

6. "plt.xlabel('Tahun')" dan "plt.ylabel('Pendapatan (dalam USD)')" berfungsi untuk menambahkan label untuk sumbu x (tahun) dan sumbu y (pendapatan dalam USD).

7. "plt.legend()" berfungsi untuk menampilkan legenda yang menjelaskan garis dan titik-titik dalam grafik.

8. "plt.grid(True)" berfungsi untuk menampilkan grid di dalam line chart.

9. "plt.show()" berfungsi untuk menampilkan grafik line chart yang telah dibuat.

Dengan menjalankan kodingan ini, akan didapatkan line chart yang menampilkan pendapatan tahunan dengan outliers yang ditandai sebagai titik-titik merah di atas line chart.
