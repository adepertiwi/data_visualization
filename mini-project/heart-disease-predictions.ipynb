# Heart Disease Predictions

```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
print("Setup Complete")
```

Code di atas digunakan untuk melakukan setup lingkungan kerja untuk visualisasi data menggunakan pustaka Python, khususnya Pandas, Matplotlib, dan Seaborn.

1. `import pandas as pd`: Ini mengimpor pustaka Pandas dan memberikan alias `pd` sehingga kita dapat menggunakan fungsi-fungsinya dengan menulis `pd.` di depannya.

2. `pd.plotting.register_matplotlib_converters()`: Ini adalah fungsi dari Pandas yang mendaftarkan konverter untuk digunakan dengan Matplotlib agar dapat mengonversi tanggal dan waktu secara otomatis saat melakukan visualisasi.

3. `import matplotlib.pyplot as plt`: Ini mengimpor modul pyplot dari pustaka Matplotlib dan memberikan alias `plt` sehingga kita dapat menggunakan fungsi-fungsinya dengan menulis `plt.` di depannya.

4. `%matplotlib inline`: Ini adalah magic command IPython yang memungkinkan plot dari Matplotlib ditampilkan secara langsung di Jupyter Notebook atau JupyterLab.

5. `import seaborn as sns`: Ini mengimpor pustaka Seaborn dan memberikan alias `sns` sehingga kita dapat menggunakan fungsi-fungsinya dengan menulis `sns.` di depannya.

6. `print("Setup Complete")`: Ini hanya mencetak pesan "Setup Complete" untuk memberi tahu bahwa setup telah selesai dilakukan.

Secara keseluruhan, code ini mempersiapkan lingkungan kerja untuk melakukan visualisasi data dengan Pandas, Matplotlib, dan Seaborn di Jupyter Notebook atau JupyterLab.

```python
# load data
heart_disease = pd.read_csv('input/heart.csv')
heart_disease.head()
```

Code di atas memuat data dari file CSV yang disebut "heart.csv" ke dalam sebuah DataFrame Pandas yang disebut "heart_disease", kemudian menampilkan lima baris pertama dari DataFrame tersebut menggunakan metode `.head()`. 

```python
# distribution of age using histogram
plt.figure(figsize=(10,6))
sns.distplot(a=heart_disease['age'], kde=False)
plt.title('Distribution of Age')
plt.show()
```

Code di atas digunakan untuk membuat histogram yang menggambarkan distribusi umur (age) dalam data penyakit jantung yang terdapat dalam DataFrame "heart_disease". 

Berikut adalah penjelasan dari setiap baris code:
1. `plt.figure(figsize=(10,6))`: Ini membuat sebuah gambar (figure) baru dengan ukuran 10x6 inci.
2. `sns.distplot(a=heart_disease['age'], kde=False)`: Ini menggunakan fungsi `distplot` dari pustaka Seaborn untuk membuat histogram distribusi umur (`age`) dari data yang ada dalam DataFrame "heart_disease". Argumen `a` digunakan untuk menyediakan data yang akan digunakan dalam plot, sedangkan `kde=False` menonaktifkan kurva kepadatan kernel yang biasanya disertakan secara default.
3. `plt.title('Distribution of Age')`: Ini menambahkan judul "Distribution of Age" ke plot.
4. `plt.show()`: Ini menampilkan plot histogram yang sudah dibuat.

Jadi, code tersebut menghasilkan histogram yang menggambarkan distribusi umur dalam data penyakit jantung.

```python
# distribution of age using KDE
plt.figure(figsize=(10,6))
sns.kdeplot(data=heart_disease['age'], shade=True)
plt.title('Distribution of Age')
plt.show()
```

Code di atas digunakan untuk membuat plot KDE (Kernel Density Estimation) yang menggambarkan distribusi umur (age) dalam data penyakit jantung yang terdapat dalam DataFrame "heart_disease". Penjelasan setiap baris code adalah sebagai berikut:

1. `plt.figure(figsize=(10,6))`: Membuat sebuah gambar (figure) baru dengan ukuran 10x6 inci.
2. `sns.kdeplot(data=heart_disease['age'], shade=True)`: Menggunakan fungsi `kdeplot` dari pustaka Seaborn untuk membuat plot KDE (Kernel Density Estimation) yang menggambarkan distribusi umur (`age`) dari data yang ada dalam DataFrame "heart_disease". Argumen `data` digunakan untuk menyediakan data yang akan digunakan dalam plot, sedangkan `shade=True` digunakan untuk mengisi area di bawah kurva kepadatan.
3. `plt.title('Distribution of Age')`: Menambahkan judul "Distribution of Age" ke plot.
4. `plt.show()`: Menampilkan plot KDE yang sudah dibuat.

Jadi, code tersebut menghasilkan plot KDE yang menggambarkan distribusi umur dalam data penyakit jantung.

```python
# 2D KDE plot
plt.figure(figsize=(10,6))
sns.jointplot(x=heart_disease['age'], y=heart_disease['chol'], kind="kde")
plt.title('Distribution of Age and Cholesterol')
plt.show()
```

Code di atas digunakan untuk membuat plot KDE (Kernel Density Estimation) dua dimensi yang menggambarkan hubungan antara umur (age) dan kadar kolesterol (chol) dalam data penyakit jantung yang terdapat dalam DataFrame "heart_disease".

Penjelasan setiap baris code adalah sebagai berikut:

1. `plt.figure(figsize=(10,6))`: Membuat sebuah gambar (figure) baru dengan ukuran 10x6 inci.
2. `sns.jointplot(x=heart_disease['age'], y=heart_disease['chol'], kind="kde")`: Menggunakan fungsi `jointplot` dari pustaka Seaborn untuk membuat plot dua dimensi yang menampilkan hubungan antara umur (`age`) dan kadar kolesterol (`chol`) dalam data yang ada dalam DataFrame "heart_disease". Argumen `x` dan `y` digunakan untuk menentukan variabel yang akan ditampilkan di sumbu x dan y, sedangkan `kind="kde"` digunakan untuk menunjukkan bahwa yang akan digunakan adalah plot KDE.
3. `plt.title('Distribution of Age and Cholesterol')`: Menambahkan judul "Distribution of Age and Cholesterol" ke plot.
4. `plt.show()`: Menampilkan plot KDE dua dimensi yang sudah dibuat.

Jadi, code tersebut menghasilkan plot KDE dua dimensi yang menunjukkan distribusi umur dan kadar kolesterol dalam data penyakit jantung.

```python
# Visualisasi korelasi antara variabel
plt.figure(figsize=(10, 6))
sns.heatmap(heart_disease.corr(), annot=True, cmap='coolwarm')
plt.title('Korelasi Antara Variabel')
plt.show()
```

Code di atas digunakan untuk membuat sebuah heatmap yang memvisualisasikan korelasi antara variabel dalam data penyakit jantung yang terdapat dalam DataFrame "heart_disease".

Penjelasan setiap baris code adalah sebagai berikut:

1. `plt.figure(figsize=(10, 6))`: Membuat sebuah gambar (figure) baru dengan ukuran 10x6 inci.
2. `sns.heatmap(heart_disease.corr(), annot=True, cmap='coolwarm')`: Menggunakan fungsi `heatmap` dari pustaka Seaborn untuk membuat heatmap korelasi antara variabel dalam DataFrame "heart_disease". Fungsi `corr()` digunakan untuk menghitung korelasi antar kolom dalam DataFrame, sedangkan argumen `annot=True` digunakan untuk menampilkan nilai korelasi di dalam setiap sel heatmap, dan `cmap='coolwarm'` digunakan untuk menentukan skema warna heatmap.
3. `plt.title('Korelasi Antara Variabel')`: Menambahkan judul "Korelasi Antara Variabel" ke plot.
4. `plt.show()`: Menampilkan heatmap korelasi yang sudah dibuat.

Jadi, code tersebut menghasilkan visualisasi korelasi antara variabel dalam data penyakit jantung menggunakan heatmap.

```python
# Menampilkan daftar nama kolom dalam DataFrame heart_data
print(heart_disease.columns)

# Visualisasi hubungan antara usia dan tekanan darah maksimal (thalach)git
plt.figure(figsize=(10, 6))
sns.scatterplot(x='age', y='thalach', data=heart_disease, hue='target', palette='coolwarm')
plt.title('Hubungan Umur dan Tekanan Darah Maksimal')
plt.xlabel('Umur')
plt.ylabel('Tekanan Darah Maksimal')
plt.show()
```

Penjelasan untuk code tersebut adalah:

1. `print(heart_disease.columns)`: Ini mencetak daftar nama kolom dalam DataFrame "heart_disease".
2. `plt.figure(figsize=(10, 6))`: Membuat sebuah gambar (figure) baru dengan ukuran 10x6 inci untuk plot selanjutnya.
3. `sns.scatterplot(x='age', y='thalach', data=heart_disease, hue='target', palette='coolwarm')`: Menggunakan fungsi `scatterplot` dari pustaka Seaborn untuk membuat scatter plot yang menunjukkan hubungan antara umur (`age`) dan tekanan darah maksimal (`thalach`) dalam data yang ada dalam DataFrame "heart_disease". Argumen `hue='target'` digunakan untuk membedakan titik-titik berdasarkan variabel target, sedangkan `palette='coolwarm'` digunakan untuk menentukan skema warna yang akan digunakan.
4. `plt.title('Hubungan Umur dan Tekanan Darah Maksimal')`: Menambahkan judul "Hubungan Umur dan Tekanan Darah Maksimal" ke plot.
5. `plt.xlabel('Umur')`: Menambahkan label sumbu x dengan teks "Umur".
6. `plt.ylabel('Tekanan Darah Maksimal')`: Menambahkan label sumbu y dengan teks "Tekanan Darah Maksimal".
7. `plt.show()`: Menampilkan scatter plot yang sudah dibuat.

Jadi, code tersebut menghasilkan visualisasi hubungan antara umur dan tekanan darah maksimal dalam data penyakit jantung, dengan pemisahan berdasarkan variabel target.
