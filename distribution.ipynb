Load data from CSV

```python
cancer = pd.read_csv('input/cancer.csv')
cancer.head()
```

Membuat dataframe menggunakan pandas

```python
df = pd.DataFrame(cancer)
```

Figure dengan ukuran 10x6 dibuat menggunakan `plt.figure(figsize=(10,6))`. Kemudian, fungsi `sns.distplot()` digunakan dari library Seaborn (`sns`) untuk membuat histogram dari 'Radius (mean)'. lalu `a=df['Radius (mean)']` digunakan untuk menentukan data yang akan digunakan untuk plot histogram. Argumen `kde=False` menghilangkan plot KDE (Kernel Density Estimate) dari histogram, sehingga hanya histogram yang akan ditampilkan di dalam figure.

```python
# Histogram
plt.figure(figsize=(10,6))
sns.distplot(a=df['Radius (mean)'], kde=False)
plt.title('Distribution of Radius (mean)')
plt.show()
```

figure dengan ukuran 10x6 dibuat menggunakan `plt.figure(figsize=(10,6))`. Kemudian, function `sns.kdeplot()` dari library Seaborn (`sns`) digunakan untuk membuat plot KDE dari  'Radius (mean)'. lalu `data=df['Radius (mean)']` digunakan untuk menentukan data yang akan digunakan untuk plot KDE. Argumen `shade=True` mengisi area di bawah kurva KDE dengan warna.

```python
# KDE plot
plt.figure(figsize=(10,6))
sns.kdeplot(data=df['Radius (mean)'], shade=True)
plt.title('Distribution of Radius (mean)')
plt.show()
```

Sebuah figure dengan ukuran 10x6 dibuat menggunakan `plt.figure(figsize=(10,6))`. Kemudian, function `sns.jointplot()` dari librarySeaborn (`sns`) digunakan untuk membuat plot KDE dua dimensi. Argumen `x=df['Radius (mean)']` dan `y=df['Texture (mean)']` menentukan data yang akan digunakan untuk sumbu-x dan sumbu-y. Argumen `kind="kde"` menentukan jenis plot yang akan digunakan, yaitu KDE.

```python
# 2D KDE plot
plt.figure(figsize=(10,6))
sns.jointplot(x=df['Radius (mean)'], y=df['Texture (mean)'], kind="kde")
plt.title('Distribution of Radius (mean) and Texture (mean)')
plt.show()
```

Sebuah figure dengan ukuran 10x6 dibuat menggunakan `plt.figure(figsize=(10,6))`. Kemudian, function `sns.jointplot()` dari library Seaborn (`sns`) digunakan untuk membuat plot scatter 2 dimemnsi. lalu `x=df['Radius (mean)']` dan `y=df['Texture (mean)']` menentukan data yang akan digunakan untuk sumbu X dan sumbu Y. Argumen `kind="scatter"` menentukan jenis plot yang akan digunakan, yaitu scatter plot.

```python
# 2D KDE plot with scatter plot
plt.figure(figsize=(10,6))
sns.jointplot(x=df['Radius (mean)'], y=df['Texture (mean)'], kind="scatter")
plt.title('Distribution of Radius (mean) and Texture (mean)')
plt.show()
```

Sebuah figure dengan ukuran 10x6 dibuat menggunakan `plt.figure(figsize=(10,6))`. Kemudian, 2 function `sns.kdeplot()` dari librarySeaborn (`sns`) digunakan untuk membuat plot KDE dari kedua kolom. Argumen `data=df['Radius (mean)']` dan `data=df['Texture (mean)']` menentukan data yang akan digunakan untuk masing-masing plot KDE. Argumen `shade=True` mengisi area di bawah kurva KDE dengan warna. Argumen `color='r'` dan `color='b'` menentukan warna untuk masing-masing plot KDE.

```python
# Color-coded plots
plt.figure(figsize=(10,6))
sns.kdeplot(data=df['Radius (mean)'], shade=True, color='r')
sns.kdeplot(data=df['Texture (mean)'], shade=True, color='b')
plt.title('Distribution of Radius (mean) and Texture (mean)')
plt.show()
```
